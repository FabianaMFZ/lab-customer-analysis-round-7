# Processing data

# Normalize (numerical)

```python
# Normalization, in the context of data preprocessing, refers to the process of rescaling features to a range of [0, 1]. 
#It's useful when the features have different ranges or units. Normalization ensures that all features are on a similar scale, preventing some features from dominating others.

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(df_numerical)
normalized_data = pd.DataFrame(normalized_data, columns=df_numerical.columns)
df_numerical = normalized_data
df_numerical
```

```python
# Standardization, also known as z-score normalization, involves rescaling features to have a mean of 0 and a standard deviation of 1. 
# This is useful when your data might not follow a normal distribution, but you still want to make features comparable. 

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
normalized_data = scaler.fit_transform(df_numerical)
normalized_data = pd.DataFrame(normalized_data, columns=df_numerical.columns)
df_numerical = normalized_data
df_numerical
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

Y = df_numerical['total_claim_amount']
X = df_numerical.drop (['total_claim_amount'], axis = 1)
lm = linear_model.LinearRegression()
lm = lm.fit(X,Y)

lm.score(X,Y)
```

```python
predictions = lm.predict(X)
predictions
```

# Model validation

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

print("R2 value is = ",round(r2_score(Y, predictions),2))
print("The intercept of the model is = ",lm.intercept_)
print("The coefficients of the model are = ",lm.coef_)

mse = mean_squared_error (Y, predictions)
mae = mean_absolute_error (Y, predictions)
print("The mse of the model is = ",mse)
print("The root mse of the model is = ",np.sqrt(mse))
print("The mean absolute error of the model is = ",mae)
```

```python
'''
R-squared (R2) value:
R-squared is a measure of how well the independent variables (features) explain the variance in the dependent variable (target) in your model.
In this case, an R2 value of 0.52 means that approximately 52% of the variance in the dependent variable is explained by the independent variable(s) included in the model.

Intercept of the model:
The intercept is the value of the dependent variable when all independent variables are zero.
In this model, the intercept is approximately 72.39.

Coefficients of the model:
Coefficients represent the slopes of the regression line for each independent variable.

Mean Squared Error (MSE):
MSE measures the average squared difference between the actual and predicted values in themodel.
In this model, the MSE is approximately 40615.74.

Root Mean Squared Error (RMSE):
RMSE is the square root of the MSE and provides an interpretable measure of the average magnitude of the errors in the model's predictions.
In this model, the RMSE is approximately 201.53.

Mean Absolute Error (MAE):
MAE measures the average absolute difference between the actual and predicted values in the model.
In this model, the MAE is approximately 144.38.
'''
```
